# Type to token ratio

Defining a function find_ttr to get the file as its parameter and provide TTR and unique tokens as output.

```python
def find_ttr(file):
    contents = file.read()
    document= re.sub(r'[^\w]', ' ', contents)
    document=document.lower()
    tokens=nlp.word_tokenize(document)
    types=nlp.Counter(tokens)
    TTR= (len(types)/len(tokens))*100
    print("TTR is: ",TTR)
    print("Number of tokens: ",len(tokens))
    print("Number of Unique tokens: ",len(set(tokens)))
    print("\n")
    print("Unique tokens are: ",set(tokens))
```

```python
def ttr(text):
    types=len(set(text))
    tokens=len(text)
    return types/tokens
```

Here, we are calculating the moving average ttr for a window of 1000 words. 

```python
def movavgTTR(text):
    counter=0
    total=0
    for i in range(0, (len(text))):
        total=total+ttr(text[i:i+999])
        counter=counter+1
    return total/counter
```

# Zipfs Law

```python
import nltk 
brown = nltk.corpus.gutenberg.raw('chesterton-brown.txt')
brown
```

```python
brown_tokens = nlp.word_tokenize(brown)
token_count = list(set(brown_tokens))
token_count
```

```python
frequency_of_words= {}
for i in (token_count):
    frequency_of_words[i] = brown_tokens.count(i)
```

```python
frequency_of_words
```

```python
sorted_frequency_of_words = sorted(frequency_of_words.items(),key=lambda x: x[1],reverse=True)
sorted_frequency_of_words
```

```python
freq_df = pd.DataFrame(sorted_frequency_of_words, columns=['type','freq'])
freq_df.sort_values('freq',ascending = False, inplace=True)
freq_df.reset_index(drop=True,inplace=True)
freq_df.index+=1
```

```python
plt.plot(np.log(freq_df.index),np.log(freq_df.freq))
plt.xlabel('Rank')
plt.ylabel('Frequency')
```

Zipf's Law states that the relative frequency of a word is inversely proportional to its rank. Here we can observe the same. In our code we took the frequency of each word and sorted them in the descending order. Initially we can see at lowest rank the frequency of word occurancy is very high. As we keep increasing the rank the frequency keep decreasing.

# Heaps law

```python
from nltk.corpus import brown
brown.categories()
brown = brown.words(categories='news')[:1000]
```

```python
types=[]
heap_tot = 100
while(heap_tot<len(brown)):
    types.append([heap_tot,len(set(brown[:heap_tot]))])
    heap_tot=heap_tot+100
```

```python
heaps_law_df= pd.DataFrame(types,columns=["token","types"])
heaps_law_df
```

```python
plt.plot(heaps_law_df.token,heaps_law_df.types)
plt.ylabel('Different types of tokens (|V|)')
plt.xlabel('Total no. of tokens (N)')
plt.show()
```

Heaps' law suggests that (i) the dictionary size continues to increase with more documents in the collection, rather than a maximum vocabulary size being reached, and (ii) the size of the dictionary is quite large for large collections.
Here we can observe that initially for 100 tokens we have around 66 types and as we increasing the tokens the types keep increasing. And finally for 900 tokens we have 401 types.
